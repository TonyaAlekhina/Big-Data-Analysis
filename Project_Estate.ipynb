Это достаточно важная характеристика, но скорее всего эта информация была проигнорирована при формировании объявления, т.к. не является обязательной. Возможно, продавец посчитал досточной информации об общей площади и числе комнат. Возможно, это квартира со свободной планировкой или квартира-студия и продавец не может определить площадь кухни. 

Пропуски: жил. площадь и площадь кухни. Пропуски по данным значениям можно заполнить на основе информации об общей площади недвижимости. Посчитаем среднюю долю жилой площади и площади кухни в общей, и заполним недостающую информацию средними значениями.

Жилую площадь заполним медианными значениями в соотвествии с числом комнат. Для этого сгруппируем данные по числу комнат и вычислим медианное значение жилой площади.


```python
living = df.groupby('rooms')['living_area'].median()
living
```

```python
for rooms_number in living.index:
    df.loc[(df['rooms'] == rooms_number), 'living_area'] = df.loc[(df['rooms'] == rooms_number),'living_area'].fillna(living[rooms_number])

# Проверим, что пропусов в этом столбце не осталось
df['living_area'].isnull().sum()
```

Площадь кухни заполним средними значениями площади кухни по всему датасету.

```python
df['kitchen_area'] = df['kitchen_area'].fillna(df['kitchen_area'].mean())
df['kitchen_area'].isnull().sum()
```

Пропущенных значений в столбце `airports_nearest`: 23.38% Пропущенных значений в столбце `cityCenters_nearest`: 23.29%

Пропуски, скорее всего, связаны с тем, что эти объекты недвижимости находятся достаточно далеко и от аэропорта, и от центра города. 


Попробуем заполнить расстояние до ближайшего аэропорта и центра города медианой в соответствии с населенными пунктами

```python
# loc_median_air - медианы расстояний до аэропорта по населенным пунктам

print('Медианы расстояний airports_nearest в locality_name : ')
loc_median_air = df.groupby('locality_name')['airports_nearest'].median().sort_values()
print(loc_median_air)
```

```python
# Функция для присваивания пропуску значения медианы
def med_insert(loc_dist):
    if loc_dist not in loc_median_air:
        return np.nan
    return loc_median_air[loc_dist]
```

```python
# Заполним пропуски в столбце airports_nearest на основании полученных значений. Используем метод apply
df.loc[df['airports_nearest'].isna() == True, 'airports_nearest'] = df['locality_name'].apply(med_insert)

#Проверим количество оставшихся пропусков
print('Всего пропущенных значений в столбце ='), df['airports_nearest'].isna().sum()
```

```python
# loc_median_center - медианы расстояний до центра города по населенным пунктам
loc_median_center = df.groupby('locality_name')['cityCenters_nearest'].median().sort_values()
print('Медианы расстояний cityCenters_nearest в locality_name : ')
print(loc_median_center)
```
